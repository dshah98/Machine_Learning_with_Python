<h1>Polynomial Regression</h1>

- y = b0 + b1X1 + b2(X1)^2 + ....... + bn(X1)^n

**Importing Libraries**

```python
# Numpy allows us to work with array.
import numpy as np

# Maptplotlib which allows us to plot some chart.
import matplotlib.pyplot as plt

# Pandas allows us to not only import the datasets but also create the matrix of features(independent) and 
# dependent variable.
import pandas as pd
```

**Importing the Dataset**

```python
dataset = pd.read_csv("Position_Salaries.csv")
```

- The independent variable usally in the first columns of dataset and dependent variable usally in the last columns of the data sets.
- X is Independent Variable.
- Y is Dependent Variable.

```python
x = dataset.iloc[:, 1:-1].values      # iloc is index of location
y = dataset.iloc[:, -1].values

print(x)
```

```python
print(y)
```

**Training the Linear Regression Model on the whole dataset**

```python
# Importing Package
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()

# Fitting
linear_reg.fit(x, y)
```

**Training the Polynomial Regression Model on the whole dataset**

```python
# Importing Package
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree = 2)

# Transforming matrix
x_poly = poly_reg.fit_transform(x)

# Polynomial Model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)
```

**Visualising the Linear Regression Model**

```python
# Real Salaries.
plt.scatter(x, y, color = "red")

# Predicted of Linear Regression Model.
plt.plot(x, linear_reg.predict(x), color = "blue")

plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

**Visualising the Polynomial Regression Model**

```python
# Real Salaries.
plt.scatter(x, y, color = "red")

# Predicted of Polynomial Regression Model.
plt.plot(x, lin_reg_2.predict(x_poly), color = "blue")

plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

- Degree value changes to 4.

```python
poly_reg = PolynomialFeatures(degree = 4)

# Transforming matrix
x_poly = poly_reg.fit_transform(x)

# Polynomial Model
lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_poly, y)

# Real Salaries.
plt.scatter(x, y, color = "red")

# Predicted of Polynomial Regression Model.
plt.plot(x, lin_reg_2.predict(x_poly), color = "blue")

plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

**Visualising the Polynomial Regression result (for higher resolution and smoother curve)**

```python
# Increaing the density
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)

# Plotting
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')

plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

**Predicting a new result with Linear Regression**

```python
linear_reg.predict([[6.5]])
```

**Predicting a new result with Polynomial Regression**

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
